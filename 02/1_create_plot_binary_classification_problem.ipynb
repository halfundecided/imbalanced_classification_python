# Chapter II: Intuition for Imbalanced Classification
Differences in the class distribution for an imbalanced classification problem will influence the choice of data preparation and modeling algorithms. Therefore it is critical that practitioners develop an intuition for the implications for different class distributions. 

## 1. Create and Plot a Binary Classification Problem
`make_blobs()` function can be used  to generate a specified number of examples from a test classification problem with a specified number of classes. 

```
from numpy import where
from matplotlib import pyplot
from sklearn.datasets import make_blobs
```

```
# Example of creating a synthetic dataset 
# generate 1,000 examples for a binary classification problem with two input variables, 0 and 1. 
X, y = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=1, cluster_std=3)
```

```
# Plotting data points by class label 
# create scatter plot for samples from each class 
for class_value in range(2):
  # get row indexes for samples with this class
  row_ix = where(y == class_value)
  # create scatter of these samples
  pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show the plot
pyplot.show()
```

There is an equal number of examples in each class, in this case, 500, and we can imagine drawing a line to resonably separate the classes.
